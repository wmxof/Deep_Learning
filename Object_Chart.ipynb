**수직 막대그래프(Bar Chart)**

불연속적일 경우: 막대 그래프

```python
import matplotlib.pyplot as plt
import numpy as np


x=np.arange(3)
values = [72,105,125]

plt.bar(x,values)
plt.show()
```

np.arange(n) : 0부터 n-1까지, 크기가 n인 1차원 배열 생성

```python
import matplotlib.pyplot as plt
import numpy as np


x=np.arange(3)
grade=['2','3','4']
values = [72,105,125]
colors=['#505467','royalblue','C4']

plt.bar(x,values,color=colors,width=0.4)
plt.xticks(x,grade)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x=np.arange(3)
grade=['2','3','4']
values = [72,105,125]
colors=['#505467','royalblue','C4']

plt.bar(x,values,color='w',align='edge',edgecolor='lightgreen',linewidth=3,tick_label=grade,width=0.6)
plt.show()
```

```python
x=np.arange(3)
grade=['2','3','4']
values = [72,105,125]
colors=['#505467','royalblue','C4']

plt.barh(x,values,color='w',align='edge',edgecolor='lightgreen',linewidth=3,tick_label=grade,height=0.6)
plt.show()
```

**산점도(Scatter)**

직교 좌표계(도표)를 이용해 좌표 상의 점들을 표시, 두 변수 간의 관계 나타내는 시각화 유형


![image.png]()

```python
np.random.seed(42)

n=30
x=np.random.rand(n)
y=np.random.rand(n)

plt.scatter(x,y)
plt.show()
```

np.random.rand(n):0부터 1까지의 값을 n만큼 할당

np.random.seed(n) : 컴퓨터에 정해져 있는 시드n 값을 찾는 함수, 즉 값을 정해주는 함수라고 생각할 수 있음

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)

n=30
x=np.random.rand(n)
y=np.random.rand(n)

area=(10*np.random.rand(n))**2
colors=np.random.rand(n)
#area : 마커 크기 설정

#cmap : 색의 종류를 지정하는 파라미터
plt.scatter(x,y,s=area,c=colors,alpha=0.5, cmap='spring')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)

n=30
x=np.arange(20)+1
y=x*x

colors=x/20.
#정규화 과정(값을 0~1로 설정)

plt.scatter(x,y,s=y,c=colors, cmap='gist_rainbow')
plt.text(0.5, 400, 'Lee SuBin')
plt.colorbar()
plt.savefig("C:/Users/PC-01/Pictures/LeeSuBin.png")
#plt.savefig() : 지정한 경로로 이미지 저장
plt.show()
```
